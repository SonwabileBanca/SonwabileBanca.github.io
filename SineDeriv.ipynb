```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x values
x = np.linspace(0, 2*np.pi, 100)

# Compute the sine function
y_sin = np.sin(x)

# Compute the first derivative of sine
y_sin_prime = np.gradient(y_sin, x)

# Compute the second derivative of sine
y_sin_double_prime = np.gradient(y_sin_prime, x)

# Plotting
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the sine function
ax.plot(x, y_sin, y_sin_prime, label='sin(x)', color='blue')

# Plot the first derivative of sine
ax.plot(x, y_sin_prime, y_sin_double_prime, label="sin'(x)", color='green')

# Plot the second derivative of sine
ax.plot(x, y_sin_double_prime, np.zeros_like(x), label="sin''(x)", color='red')

# Add legend
ax.legend()

# Add title and labels
ax.set_title('Sine Function and Its Derivatives')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Show plot
plt.show()
```
