# Signal Processing: Hann Window

A square pulse is a signal that abruptly transitions between two levels. When directly analyzed in the frequency domain, it can introduce significant spectral leakage, which can obscure the true frequency content of the signal.


A Hann window is a mathematical function used to gradually taper the edges of a signal, reducing the impact of spectral leakage. By applying a Hann window to a square pulse, we can smooth the edges and obtain a more accurate frequency spectrum.

### **This programme was done using numpy only**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
pulse_duration = 100
amplitude = 1
sampling_rate = 1000

# Create the square pulse
t = np.arange(pulse_duration) / sampling_rate
square_pulse = np.zeros_like(t)
square_pulse[int(pulse_duration/2):] = amplitude

# Create the Hann window
hann_window = 0.5 * (1 - np.cos(2 * np.pi * np.arange(pulse_duration) / (pulse_duration - 1)))

# Apply the Hann window to the square pulse
smoothed_pulse = square_pulse * hann_window

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, square_pulse, label='Square Pulse')
plt.plot(t, hann_window, label='Hann Window')
plt.plot(t, smoothed_pulse, label='Smoothed Pulse')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

**This code was done using Scipy and signal libraries**

```python
from scipy import signal
sig = np.repeat([0., 1., 0.], 100)
win = signal.windows.hann(50)
filtered = signal.convolve(sig, win, mode='same') / sum(win)
```

```python
import matplotlib.pyplot as plt
fig, (ax_orig, ax_win, ax_filt) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.set_title('Original pulse')
ax_orig.margins(0, 0.1)
ax_win.plot(win)
ax_win.set_title('Filter impulse response')
ax_win.margins(0, 0.1)
ax_filt.plot(filtered)
ax_filt.set_title('Filtered signal')
ax_filt.margins(0, 0.1)
fig.tight_layout()
fig.show()
```
