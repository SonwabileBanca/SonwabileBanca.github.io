```python
#Now create a for loop for multiple 3x3 matrices input for better understanding


def eval_detA(A):
  """Calculates the determinant of a nxn matrix A using TensorFlow.

  Args:
    A: A nxn NumPy array.

  Returns:
    The determinant of A.
  """

  # Convert the NumPy array to a TensorFlow tensor
  A_tensor = tf.convert_to_tensor(A, dtype=tf.float32)

  # Calculate the determinant using TensorFlow's `linalg.det` function
  determinant = tf.linalg.det(A_tensor)

  return determinant


matrices = [  #Matrices
    np.array([
        [11, 3, 0.5],
        [-5, 2, 0],
        [-2, 6, -1]
    ]),
    np.array([
        [2, 9, 6],
        [4, -3, 7],
        [-6, 1, 0]
    ]),
    np.array([  #Inverse matrix
        [1, 0, 0],
        [0, 1, 0],
        [0, 0, 1]
    ])
]

for i, A in enumerate(matrices):
  result = eval_detA(A)
  print(f"Determinant of Matrix {i+1}:", result.numpy())

  # Visualize the matrix A
  plt.imshow(A, cmap='viridis', interpolation='nearest')
  plt.colorbar()
  plt.title(f"Matrix {i+1}")
  plt.show()
```
