```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def f(t, y):
    return -2*t*y**2 + 1

# Set initial conditions
t0 = 0
y0 = np.array([0]) # y0 should be a 1-dimensional array

# Set time span for the solution
t_span = [t0, 10]  # Adjust the end time as needed

# Solve the IVP using RK4 method
sol = solve_ivp(f, t_span, y0, method='RK45')  # You can choose other methods like 'RK23'

# Extract the solution
t_sol = sol.t
y_sol = sol.y[0]

# Plot the solution
plt.plot(t_sol, y_sol)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solution of y\' = -2xy^2 + 1')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'Time': t_sol, 'Solution': y_sol})
```
